```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.array([1, 2, 3, 4, 5])  # X values
y = np.array([6, 7, 8, 9, 10])  # Y values
a = np.array([0.8, 1.2, 1.4, 0.9, 1.1])  # A values
b = np.array([1.1, 1.3, 0.7, 1.5, 0.9])  # B values
c = np.array([1.2, 0.9, 1.1, 1.3, 0.8])  # C values

# Calculate the smallest value among A, B, and C for each (X, Y) point
z = np.minimum.reduce([a, b, c])

# Create a scatter plot with color-coded points
plt.scatter(x, y, c=z, cmap='coolwarm', vmin=np.min(z), vmax=np.max(z), edgecolors='black')
plt.colorbar(label='Minimum Value (A, B, C)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Heatmap')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.array([0, 1, 2, 3, 4])  # X values
y = np.array([0, 1, 2, 3, 4])  # Y values
a = np.array([0.8, 1.2, 1.4, 0.9, 1.1])  # A values
b = np.array([1.1, 1.3, 0.7, 1.5, 0.9])  # B values
c = np.array([1.2, 0.9, 1.1, 1.3, 0.8])  # C values

# Determine the index of the smallest value among A, B, and C for each point
min_indices = np.argmin([a, b, c], axis=0)

# Create a scatter plot with color-coded markers
colors = ['yellow', 'blue', 'red']
for i in range(len(x)):
    plt.scatter(x[i], y[i], color=colors[min_indices[i]], edgecolors='black')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Smallest Values (A, B, C)')
plt.xlim([min(x) - 1, max(x) + 1])
plt.ylim([min(y) - 1, max(y) + 1])

plt.show()
```
